```python
ip=food.drop(['Suitable_for_Diabetes'],axis=1)
op=food.Suitable_for_Diabetes
```

```python
from keras.utils import to_categorical
op = to_categorical(op,3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.15)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
#input layer
model.add(Dense(50,input_dim=12,activation='relu'))
#hidden layers
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
#output layer
model.add(Dense(3,activation='softmax'))
model.compile(Adam(learning_rate=0.01),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
h = model.fit(x_train,y_train,epochs=20,validation_data=(x_test,y_test))
```
